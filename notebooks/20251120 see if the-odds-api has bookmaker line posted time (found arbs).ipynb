````python
"""
FULL API RESPONSE ANALYSIS
Show ALL bookmakers and their timestamps
"""

import requests
import json
import os
from dotenv import load_dotenv
import ssl
import urllib3
import pandas as pd

# Load environment variables from api_setup directory
from pathlib import Path
env_path = Path('../api_setup/.env')
load_dotenv(dotenv_path=env_path)

# SSL fix
ssl._create_default_https_context = ssl._create_unverified_context
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

API_KEY = os.getenv('ODDS_API_KEY')
BASE_URL = 'https://api.the-odds-api.com/v4'

print(f"API Key loaded: {'Yes ‚úì' if API_KEY else 'No ‚úó'}")
print(f".env path: {env_path.absolute()}")

# REAL EVENT from opening day 2024-10-22 (Knicks @ Celtics)
EVENT_ID = '92cf7db605a734b10e69eabf56e1eac9'
DATE = '2024-10-22T15:00:00Z'  # 10 AM ET - our snapshot time

print(f"{'='*80}")
print(f"FULL API RESPONSE ANALYSIS")
print(f"{'='*80}")
print(f"\nEvent ID: {EVENT_ID}")
print(f"Requested snapshot time: {DATE}")
print(f"\n{'='*80}\n")

params = {
    'api_key': API_KEY,
    'regions': 'us',
    'markets': 'player_threes',
    'oddsFormat': 'american',
    'dateFormat': 'iso',
    'date': DATE
}

try:
    response = requests.get(
        f"{BASE_URL}/historical/sports/basketball_nba/events/{EVENT_ID}/odds",
        params=params,
        verify=False
    )
    response.raise_for_status()
    
    data = response.json()
    
    print("="*80)
    print("TOP-LEVEL RESPONSE FIELDS")
    print("="*80)
    for key in data.keys():
        print(f"  - {key}")
    
    print(f"\n{'='*80}")
    print("EVENT DATA")
    print("="*80)
    event_data = data.get('data', {})
    print(f"\nGame: {event_data.get('away_team')} @ {event_data.get('home_team')}")
    print(f"Game starts: {event_data.get('commence_time')}")
    print(f"Sport: {event_data.get('sport_title')}")
    
    # BOOKMAKER TIMESTAMP ANALYSIS
    print(f"\n{'='*80}")
    print("ALL BOOKMAKERS - TIMESTAMP ANALYSIS")
    print("="*80)
    
    bookmakers = event_data.get('bookmakers', [])
    print(f"\nTotal bookmakers: {len(bookmakers)}")
    
    timestamp_data = []
    
    for i, bookmaker in enumerate(bookmakers, 1):
        book_key = bookmaker.get('key')
        book_title = bookmaker.get('title')
        book_last_update = bookmaker.get('last_update')
        
        # Get market timestamp too
        markets = bookmaker.get('markets', [])
        market_last_update = markets[0].get('last_update') if markets else None
        num_props = len(markets[0].get('outcomes', [])) if markets else 0
        
        print(f"\n{i}. {book_title} ({book_key})")
        print(f"   Bookmaker last_update: {book_last_update}")
        print(f"   Market last_update:    {market_last_update}")
        print(f"   Number of props:       {num_props // 2}")  # Divide by 2 (over/under)
        
        timestamp_data.append({
            'bookmaker': book_title,
            'key': book_key,
            'bookmaker_last_update': book_last_update,
            'market_last_update': market_last_update,
            'num_props': num_props // 2
        })
    
    # Create DataFrame for easier analysis
    df = pd.DataFrame(timestamp_data)
    df['bookmaker_last_update'] = pd.to_datetime(df['bookmaker_last_update'])
    df['market_last_update'] = pd.to_datetime(df['market_last_update'])
    
    # Calculate time difference from our snapshot
    snapshot_time = pd.to_datetime(DATE)
    df['minutes_before_snapshot'] = (snapshot_time - df['market_last_update']).dt.total_seconds() / 60
    
    print(f"\n{'='*80}")
    print("TIMESTAMP FRESHNESS (sorted by stalest to freshest)")
    print("="*80)
    df_sorted = df.sort_values('minutes_before_snapshot', ascending=False)
    print(f"\n{df_sorted[['bookmaker', 'market_last_update', 'minutes_before_snapshot', 'num_props']].to_string(index=False)}")
    
    print(f"\n{'='*80}")
    print("KEY INSIGHTS")
    print("="*80)
    print(f"\nSnapshot requested:     {DATE}")
    print(f"Stalest bookmaker:      {df_sorted.iloc[0]['bookmaker']} ({df_sorted.iloc[0]['minutes_before_snapshot']:.1f} min old)")
    print(f"Freshest bookmaker:     {df_sorted.iloc[-1]['bookmaker']} ({df_sorted.iloc[-1]['minutes_before_snapshot']:.1f} min old)")
    print(f"Average staleness:      {df['minutes_before_snapshot'].mean():.1f} minutes")
    print(f"Max time difference:    {df['minutes_before_snapshot'].max() - df['minutes_before_snapshot'].min():.1f} minutes")
    
    print(f"\n{'='*80}")
    print("CREDITS USED")
    print("="*80)
    print(f"Cost: {response.headers.get('x-requests-last', 'N/A')} credits")
    print(f"Remaining: {response.headers.get('x-requests-remaining', 'N/A')} credits")
    
    # Save to CSV
    from pathlib import Path
    from datetime import datetime
    
    tmp_dir = Path('../tmp')
    tmp_dir.mkdir(exist_ok=True)
    
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    csv_file = tmp_dir / f'bookmaker_timestamps_{timestamp}.csv'
    
    df_sorted.to_csv(csv_file, index=False)
    
    print(f"\n{'='*80}")
    print("SAVED DATA")
    print("="*80)
    print(f"\nüíæ Saved to: {csv_file}")
    print(f"Rows: {len(df_sorted)}")
    
except Exception as e:
    print(f"Error: {e}")
    import traceback
    traceback.print_exc()
````

````python
"""
TEST TIMESTAMP CAPTURE IN NOTEBOOK
Verify that new CSV columns are being added
"""

import sys
from pathlib import Path

# Add api_setup to path (for notebooks, use relative path)
api_setup_path = Path('../api_setup').resolve()
if str(api_setup_path) not in sys.path:
    sys.path.insert(0, str(api_setup_path))

from fetch_historical_props import get_historical_events, get_historical_event_odds, parse_player_props
import pandas as pd

# Test with opening day
TEST_DATE = '2024-10-22'

print("="*80)
print("TESTING TIMESTAMP CAPTURE")
print("="*80)
print(f"\nFetching events for {TEST_DATE}...")

# Get one event
events = get_historical_events(TEST_DATE)
if not events:
    print("‚ùå No events found")
else:
    event = events[0]
    event_id = event['id']
    game_info = f"{event['away_team']} @ {event['home_team']}"
    
    print(f"‚úì Found event: {game_info}")
    print(f"  Event ID: {event_id}")
    
    # Fetch odds for this event
    print(f"\nFetching odds...")
    odds_data = get_historical_event_odds('basketball_nba', event_id, TEST_DATE, markets='player_threes')
    
    if not odds_data:
        print("‚ùå No odds data returned")
    else:
        print(f"‚úì Got odds data")
        
        # Parse props
        props = parse_player_props(odds_data)
        print(f"‚úì Parsed {len(props)} prop rows")
        
        # Convert to DataFrame
        df = pd.DataFrame(props)
        
        print(f"\n{'='*80}")
        print("CSV COLUMNS")
        print("="*80)
        print("\nColumns in DataFrame:")
        for col in df.columns:
            print(f"  ‚úì {col}")
        
        # Check for timestamp columns
        has_bookmaker_ts = 'bookmaker_last_update' in df.columns
        has_market_ts = 'market_last_update' in df.columns
        
        print(f"\n{'='*80}")
        print("TIMESTAMP VERIFICATION")
        print("="*80)
        print(f"\nbookmaker_last_update column: {'‚úì YES' if has_bookmaker_ts else '‚úó NO'}")
        print(f"market_last_update column:    {'‚úì YES' if has_market_ts else '‚úó NO'}")
        
        if has_market_ts:
            print(f"\nSample timestamps:")
            sample = df[['player', 'bookmaker', 'market_last_update']].head(5)
            print(sample.to_string(index=False))
            
            # Show unique bookmakers and their timestamps
            print(f"\n{'='*80}")
            print("BOOKMAKER TIMESTAMP ANALYSIS")
            print("="*80)
            unique_books = df.groupby('bookmaker')['market_last_update'].first().sort_values()
            print("\nBookmaker update times:")
            for book, ts in unique_books.items():
                print(f"  {book:15s} {ts}")
            
            print(f"\n‚úÖ SUCCESS - Timestamps are being captured!")
        else:
            print(f"\n‚ùå FAIL - Timestamps not in data")
        
        print(f"\n{'='*80}")
````
